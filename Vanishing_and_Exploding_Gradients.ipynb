```python
from scipy import stats

def RELU_variant_input():
    return np.arange(-5,5, 0.1) 

def plot_RELU_variant(variant_out, name, label, ax):
    layer_out = RELU_variant_input()

    _= ax.plot(layer_out, variant_out )
    _= ax.set_xlabel("Input", size=14)
    _= ax.set_ylabel(label, size=14)
    _= ax.set_title(name, size=20)

    return fig, ax

def plot_RELU_variants():
    variant_input = RELU_variant_input() 

    fig, axs = plt.subplots( 1, 3, figsize=(12,4))

    # ReLU
    _= plot_RELU_variant( variant_input * (variant_input >= 0) + 0 * (variant_input < 0),
                          "ReLU",
                          f"Output",
                          axs[0]
                         )

    # Leaky ReLU
    leakage_rate = 0.05
    _= plot_RELU_variant( variant_input * (variant_input >= 0) + variant_input * leakage_rate * (variant_input < 0),
                          "Leaky ReLU",
                          f"Output: leakage {leakage_rate}",
                          axs[1]
                         )

    # GELU
    _= plot_RELU_variant( variant_input * stats.norm.cdf(variant_input),
                          "GELU",
                          f"Output",
                          axs[2]
                         )

    fig.tight_layout()
```

# Vanishing/exploding gradients

The module on [Back propagation](Training_Neural_Network_Backprop.ipynb)
prepared us to understand the *mechanics* of the backward pass.

We now try to understand some mathematical properties of the process.

Absent this understanding, training of deep (many layer) Neural Networks was challenging for many years.

This is one reason that 
the evolution of Deep Learning
stalled for many years.

Let's summarize back propagation up until this point
- We compute the loss gradient $\loss'_\llp = \frac{\partial \loss}{\partial \y_\llp}$ of each layer $\ll$ in descending order

- The backward step  to compute the loss gradient of the preceding layer is:  
    - $\loss'_{(\ll-1)} =  \loss'_\llp \frac{\partial \y_\llp}{\partial \y_{(\ll-1)}}$

When we derived back propagation, we didn't look "inside" of the "local gradient " $\frac{\partial \y_\llp}{\partial \y_{(\ll-1)}}$




Let's look more deeply into the  term $\frac{\partial \y_\llp}{\partial \y_{(i-1)}}$

$$
\begin{array}[lllll] \\
\frac{\partial \y_\llp}{\partial \y_{(\ll-1)}} & = & \frac{\partial a_\llp ( f_\llp(\y_{(\ll-1)}, \W_\llp))}{\partial \y_{(\ll-1)}} & (\text{def. of } \y_\llp) \\
                                      & = & \frac{\partial a_\llp ( f_\llp(\y_{(\ll-1)}, W_\llp) )}{\partial f_\llp(\y_{(\ll-1)}, \W_\llp)} \frac{\partial f_\llp(\y_{(\ll-1)}, \W_\llp)}{\partial \y_{(\ll-1)}} &  (\text{chain rule}) \\
                                      & = a'_\llp f'_\llp
\end{array}
$$

where we define

$$
\begin{array}[lll] \\
a'_\llp & = & \frac{\partial a_\llp ( f_\llp(\y_{(\ll-1)}, \W_\llp) )}{\partial f_\llp(\y_{(\ll-1)}, \W_\llp)}  & \text{derivative of } a_\llp(\ldots) \text{ wrt } f_\llp(\ldots)\\
f'_\llp & = & \frac{\partial f_\llp(\y_{(\ll-1)}, W_\llp)}{\partial \y_{(\ll-1)}} & \text{derivative of } f_\llp(\ldots) \text{ wrt } \y_{(\ll-1)}\\
\end{array}
$$

$a'_\llp$ is the derivative of activation function $a_\llp$ with respect to its input.
- we saw graphs of this when we introduced activation functions
- see below

We won't explicitly write it out other than to **observe** $a'_\llp \in [0,1]$.

Substituting the value of the local gradient $\frac{\partial \y_\llp}{\partial \y_{(\ll-1)}}$ into the backward update rule for layer $\ll-1$

$$
\begin{array}[llll]\\
\loss'_{(\ll-1)} & = &  \loss'_\llp \frac{\partial \y_\llp}{\partial \y_{(\ll-1)}} \\
         & = &  \loss'_\llp a'_\llp f'_\llp
\end{array}
$$

We now do the same substitution of the value of the local gradient
for the "upstream" (layer $\ll$) loss gradient $\loss'_\llp$ 
$$
\begin{array}[llll]\\
\loss'_\llp & = & \loss'_{(\ll+1)}\frac{\partial \y_{(\ll+1)}}{\partial \y_\llp} \\
        & = &  \loss'_{(\ll+1)} a'_{(\ll+1)} f'_{(\ll+1)} 
\end{array}
$$

Substituting the expanded $\loss'_\llp$ back into $\loss'_{(\ll-1)}$
\begin{array}[llll]\\
\loss'_{(\ll-1)} & = & \loss'_{(\ll+1)} a'_{(\ll+1)} f'_{(\ll+1)} a'_\llp f'_\llp \\
& = & (a'_{(\ll+1)} a'_\llp ) \, (f'_{(\ll+1)} f'_\llp) \,\loss'_{(\ll+1)} & \text{re-arranging terms}
\end{array}

Notice the product of the derivatives of the activation function
$$
(a'_{(\ll+1)} a'_\llp )
$$

If we continue this expansion to $\loss'_{(\ll+1)}$ and beyond to final layer $L$

- we can generalize the loss gradient $\loss'_\llp$ of layer $\ll$ in terms of *all deeper layers*

$$\loss'_\llp  =   \loss'_{(L+1)} \prod_{l'=\ll+1}^L  a'_{(l')} f'_{(l')}$$

The issue is that, since 
$$
0 \le a'_\llp \le \max{z} a'_\llp(z)
$$

the product 
$$\prod_{l'=\ll+1}^L {a'_{(l')}}
$$
can be increasingly small as the number of layers $L$ grows
- if $\max{z} a'_\llp(z) < 1$.

A plot of activation functions and their derivatives will show this to be the case for many activation functions.

```python
fig, _ = nnh.plot_activations( np.arange(-5,5, 0.1) )
```

Thus, unless offset by the $f'_\llp$ terms, $\loss'_\llp$ will diminish as $L$ decreases
- and the loss gradient $\loss'_\llp$ of shallow layers $ll$ (closer to the input)
- numerically approaches 0

Should this occur, 
- the Neural Network can't learn (i.e., update weights) 
- for shallow layers 

since

$$
\begin{array}[lll] \\
\frac{\partial \loss}{\partial W_\llp} & = & \frac{\partial \loss}{\partial \y_\llp} \frac{\partial \y_\llp}{\partial W_\llp} & = & \loss'_\llp \frac{\partial \y_\llp}{\partial W_\llp}
\end{array}
$$
will approach $0$.


Let's revisit the  previous example
- of the **effect of activation function choice** on Loss and Accuracy
of a binary classifier

For each training example
- we compute the per example loss gradient
    - it is a vector (one derivative per output) 
    - so we average it across all elements
- at each layer $\ll$

By doing this for all training examples, we thus obtain a
- distribution of loss gradients
- by layer

**for each epoch** of training

Let's look at the Loss Gradient Distribution by Layer for each activation function
- as we vary the Epoch (**horizontal axis of graph**)

Focusing on Sigmoid column (**middle column**) of the chart below
- we see that mean gradients (**first row**) are essentially 0 
    - after the first epoch
    - for **non-Classifier layers**
- with very small standard deviation (**second row**): 

So *almost all samples* from the distribution of Loss Gradients are near $0$.

This illustrates
- how the Sigmoid activation
- is exposed to the Vanishing Gradient problem

and is the reason for the failure to train the NN with Sigmoid as the activation function

For the **other activation functions** (ReLU or tanh)
- the slightly larger (although still small) mean
- is offset by a larger variance

so that many gradients from the distribution are non-zero
- and the NN with these activations *can* learn



We can now diagnose one reason that training of early Deep Learning networks was difficult
- use of sigmoid activations were common (inspired by biology)
- the maximum of the derivative of the sigmoid is much smaller than $1$

The end result was that it was difficult to train **deep** Neural Networks.

## Why the ReLU became the common activation function

One half of the domain of the ReLU has non-zero (and, depending on slope, greater than 1) gradient.
- This makes it less susceptible to the Vanishing gradient problem.

Still: half of the domain of the ReLU has zero gradient, which may impede learning.

There are variants of ReLU
- constructed such that gradient is non-zero everywhere
- hopefully facilitating faster convergence in training

Some examples (and their plots below)
- Leaky ReLU
$$
\text{LeakyReLU}(x) = x * (x \ge 0) + x * \alpha * (x \lt 0)
$$
for small $\alpha$ (leakage) rate
- [Gaussian Error Linear Unit (GELU)](https://arxiv.org/pdf/1606.08415v3.pdf)
$$
\text{GELU}(x) = x * \text{NormCDF}(x)
$$
The belief is that this may facilitate learning.

# Conclusion

Something seemingly as simple as taking derivatives turned out to have some important subtleties.

The problem of gradients either shrinking to zero or growing too large is a real problem
- It can still hinder the use of very deep (many layers) networks
- This is particularly a problem in Recurrent networks
    - The depth of the "unrolled loop" is the length of the input sequence

We will explore techniques to manage the issue of vanishing and exploding gradients.
